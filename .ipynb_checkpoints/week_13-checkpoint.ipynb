```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('carseats.csv').drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

## Preprocessing the data, do the followings
 - convert the columns of 'Urban' and 'US' into 0/1
 - convert the column of ShelveLoc into {0,1,2}
 - convert the column of Sales into {0,1} with the criteria: 1 if larger than 8, 0 otherwise

```python
df.Urban = df.Urban.map({'No':0, 'Yes':1})
df.US = df.US.map({'No':0, 'Yes':1})
```

```python
df.head()
```

```python
df.ShelveLoc = pd.factorize(df.ShelveLoc)[0]
```

```python
df.head()
```

```python
df['High'] = df.Sales.map(lambda x: 1 if x>8 else 0)
```

```python
df.head()
```

## Define the data: excluding 'Sales' and 'High', everything else is input X. The column 'High' is the output Y. 

```python
X = df.drop(['Sales', 'High'], axis=1)
y = df.High
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

## The data are split into the training and test subsets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
# Training the model

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

### Plot the tree

```python
from sklearn.tree._export import plot_tree
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
plot_tree(clf, feature_names = X.columns, rounded = True, filled=True)
plt.show()
```

### Make prediction for the test data and investigate performance of the model

```python
pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```

## The tree model can be trained with different loss function
 - Gini index (Eq. 8.6)
 - entropy (Eq. 8.7)

```python
# Use the Gini index

clf1 = DecisionTreeClassifier(criterion='gini')
clf1.fit(X_train, y_train)
pred1 = clf1.predict(X_test)
```

```python
print(classification_report(y_test, pred1))
```

```python
# Use the entropy

clf2 = DecisionTreeClassifier(criterion='entropy')
clf2.fit(X_train, y_train)
pred2 = clf2.predict(X_test)
```

```python
print(classification_report(y_test, pred2))
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,pred2))
```

### From above scores, using Gini index seems a better option for this particular case.

## Next, we shall see how the accuracy of tree classification changes when the tree size is tuned. Please have a look at Fig. 8.5, which contains the essence of pruning.

```python

max_depth = []
acc_gini = []
acc_entropy = []

for i in range(1,50):
    dtree = DecisionTreeClassifier(criterion='gini', max_depth=i)
    dtree.fit(X_train, y_train)
    pred = dtree.predict(X_test)
    acc_gini.append(accuracy_score(y_test, pred))
    ####
    dtree = DecisionTreeClassifier(criterion='entropy', max_depth=i)
    dtree.fit(X_train, y_train)
    pred = dtree.predict(X_test)
    acc_entropy.append(accuracy_score(y_test, pred))
    ####
    max_depth.append(i)

d = pd.DataFrame({'acc_gini':pd.Series(acc_gini), 'acc_entropy':pd.Series(acc_entropy), 'max_depth':pd.Series(max_depth)})

# visualizing changes in parameters
fig, ax = plt.subplots(figsize=(16, 12))
plt.plot('max_depth','acc_gini', data=d, label='gini')
plt.plot('max_depth','acc_entropy', data=d, label='entropy')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend()
```

## Here are a few take-home messages
 - The accuracy grows as the tree size grows from very small values 
 - For Gini trees, the trees start to have degrading performance when size > 8
 - For entropy trees, the performance do not further improve when size > 15
 - More complex models tend to have smaller training error (you can additionally plot this information), but usually have poor test accuracy (the result above)
